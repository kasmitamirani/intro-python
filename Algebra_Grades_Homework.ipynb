```python
# Using the following information about his class, calculate the following:
# a) %age of students with a passing grade
# b) %age of women with a passing grade
# c) Average hours of study for all students
# d) Average hours of study for students with a passing grade

#****************************It has been assumed that "D" is the LOWEST passing grade****************************
```

```python
import pandas as pd
Location = r"datasets/algebradata.csv"
df = pd.read_csv(Location, header=None)
```

```python
df.columns = ['First Name','Last Name','Sex','Grade','Hours studied']
df.drop(0)
```

## Percentage of students with a passing grade

```python
# a) %age of students with a passing grade
#It has been assumed that "D" is the LOWEST passing grade
#A less sophisticated version of what will be done again below.
GradeA = df.loc[df['Grade']=='A'].count()
GradeB = df.loc[df['Grade']=='B'].count()
GradeC = df.loc[df['Grade']=='C'].count()
GradeD = df.loc[df['Grade']=='D'].count()
Passed = GradeA + GradeB + GradeC + GradeD

GradeE = df.loc[df['Grade']=='E'].count()
GradeF = df.loc[df['Grade']=='F'].count()
Failed = GradeE + GradeF

Total = Passed + Failed
PercentPass = Passed / Total
print PercentPass
```

```python
#A more sophisticated version of the same calculations.
passing = df.ix[(df['Grade']=="A") | (df['Grade']=="B") | (df['Grade']=="C") | (df['Grade']=="D")]
passnum = len(passing) * 100.0
#Without the "100" we find that 916/1000 yields 0. 100.0 exists to allow for a decimal place.
totalnum = len(df)
print "The percentage of people who passed this class with a D or higher is:"
print(passnum / totalnum)
```

## Percentage of students who are female, with a passing grade.

```python
female = df['Sex'] == "F"
```

```python
femlength = len(df.ix[df['Sex']=="F"])
print femlength
```

```python
passfemale = len(df.ix[((df['Grade']=="A") | (df['Grade']=="B") | (df['Grade']=="C") | (df['Grade']=="D")) & (df['Sex']== "F")])
print "The percentage of women who passed this class with a D or higher is:"
print (passfemale)* 100.0 / femlength
print "Note that we used the total number of females as a total, not the number of students in the dataset."
```

## Average hours of study for all students.

```python
print "The average hours of study for all students is:"
#Coercion used in order to force strings to parse anyway
print pd.to_numeric(df['Hours studied'], errors='coerce').mean()
```

## Average hours of study for all students who passed Algebra.

```python
print "The average hour of study for students who passed Algebra is:"
#returns conditions and then for specified column
print pd.to_numeric(df.ix[((df['Grade']=="A") | (df['Grade']=="B") | (df['Grade']=="C") | (df['Grade']=="D"))]['Hours studied'], errors="coerce").mean()
```
